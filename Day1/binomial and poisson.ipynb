```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
#by adding this we dont need to writen plt.show everytime
%matplotlib inline 
```

```python
#Binomial Distribition

#no. of trials
n=7
#probabilty of success
p=0.6
```

```python
#generating an array, value of x

k=np.arange(0,8)
k
```

```python
binomial= stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
#o- defines point and a line plot, by default line plot is formed
plt.plot(k, binomial, 'o-', color='m' )
plt.title("Binomial: n=%i, p=%.2f" %(n, p), fontsize=15, color='m' )
plt.xlabel("Number Of Successes")
plt.ylabel("Probability of Success")
plt.show()
```

```python
#Poisson Distribuition
lam=3
n=np.arange(0,12)
poisson= stats.poisson.pmf(n,lam)
```

```python
poisson
```

```python
plt.plot(poisson, 'o-', color='m' )
#plt.title("Poisson: Lambda", %rate , fontsize=15, color='m' )
plt.xlabel("Number Of Customers Arriving in a minute")
plt.ylabel("Probability of number of customers arriving in a minute")
plt.show()
```
