```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer= pd.read_csv("Breast Cancer Classification Data.csv")
```

```python
cancer.head()
```

```python
cancer=cancer.drop("id", axis=1)
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
plt.scatter(cancer['radius_mean'], cancer['perimeter_mean'], color= 'red')
```

```python
#create decision tree
#split the data into train and test
from sklearn.cross_validation import train_test_split
```

```python
x= cancer.drop('diagnosis', axis=1)
y= cancer['diagnosis']
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x,y, test_size=0.30, random_state=123)
```

```python
#import the cla.ssifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import  RandomForestClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
#Train the model
model_dtree= dtree.fit(xtrain, ytrain)
```

```python
#predict the unknown dataset

pred_dtree=dtree.predict(xtest)
```

```python
b=pd.crosstab(ytest, pred_dtree)
b
```

```python
acc= (101+63)/len(xtest)
acc
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(ytest, pred_dtree))  
print(classification_report(ytest, pred_dtree))  
```

```python
# lets run random forest
rf = RandomForestClassifier()
```

```python
rf_model= rf.fit(xtrain,ytrain)

rf_pred=rf.predict(xtest)
```

```python
print(confusion_matrix(ytest, rf_pred))  
print(classification_report(ytest, rf_pred)) 
```

```python
petrol.head()
```

```python
petrol.info()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import  RandomForestRegressor
```

```python
xp= petrol.drop("Petrol_Consumption", axis=1)
```

```python
yp=petrol["Petrol_Consumption"]
```

```python
xptrain, xptest, yptrain, yptest= train_test_split(xp,yp, test_size=0.30, random_state=1)
```

```python
dtree_reg= DecisionTreeRegressor()
```

```python
dtr_model= dtree_reg.fit(xptrain,yptrain)
```

```python
dtr_pred=dtr_model.predict(xptest)
```

```python
rmse= np.sqrt(np.sum((dtr_pred-yptest)**2/len(yptest)))
rmse
```

```python
mape=(abs((yptest-dtr_pred)/yptest))/len(yptest)
```

```python
sum(mape)
```

```python
rfr= RandomForestRegressor()
```

```python
rfr_model= rfr.fit(xptrain,yptrain)
```

```python
rfr_pred= rfr.predict(xptest)
```

```python
rmse= np.sqrt(np.sum((rfr_pred-yptest)**2/len(yptest)))
rmse
```

```python
mape=(np.absolute((yptest-rfr_pred)/yptest))/len(yptest)
```
