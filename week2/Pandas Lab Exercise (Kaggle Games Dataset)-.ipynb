## Data Visualization using Seaborn

### Import the seaborn library and set color codes as true

```python
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
import matplotlib.pyplot as plt
```

### Drop na values for negating issues during visualization

```python
games2= games.dropna()
games2
```

### View the distance plot for minage

```python

sns.distplot(games2['minage'])
```

### Is there a linear relationship between Minage & average_rating?

```python
sns.jointplot(games2['minage'], games2['average_rating'])

#There is no linear relationship between minage and average rating as the points scattered for a line which will be parallel to
# Yaxis, which means for different values of average_rating we will have similar values of minage or we can say minage doesnt 
#change with change in average_rating
```

### Compare the relationship between playingtime , minage and average rating using pairplot

```python
sns.pairplot(games2[['playingtime', 'minage', 'average_rating']])
```

all of the graphs donot show a linear realtionship. values on one axis are constant for different values on the other axis.

### Compare type of game and playingtime using a stripplot

```python
sns.stripplot(games2['type'], games2['playingtime'], jitter=True) 
```

Board games are played more than boardgameexpansion. board games have some similar values and two of them are extreme equal to 60000.
